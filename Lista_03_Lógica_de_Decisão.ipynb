## **Exercício 6: operações aritméticas com variáveis obtidas por classificação**

* ≤ 100 → 0.5 por kWh
* < 300 → 0.75 por kWh
* 300 → 1.0 por kWh

sendo que o cliente gastou 97 kWh no primeiro mês, 274 no segundo e 300 no terceiro.

````python
# Entre com seu código aqui
cliente = {
    "1º mês": 97,
    "2º mês": 274,
    "3º mês": 300
}

for mes, kwh in cliente.items():
  if (kwh <= 100):
    consumo = kwh * 0.5
    print(f"{mes}: {consumo}")
  elif (kwh < 300):
    consumo = kwh * 0.75
    print(f"{mes}: {consumo}")
  else:
    consumo = kwh * 1.0
    print(f"{mes}: {consumo}")

````

## **Exercício 7: operações matemáticas com if/elif**

As variáveis `x`, `y` e `operacao` definem uma operação:

* `"soma"`
* `"subtracao"`
* `"multiplicacao"`
* `"divisao"`

Se for divisão por zero, o resultado será `"erro"`.

Calcule as seguintes operações:

* `x` = 111, `y` = 4.7, `operacao` = "soma"
* `x` = 11, `y` = 44, `operacao` = "subtracao"
* `x` = 71, `y` = 4, `operacao` = "multiplicacao"
* `x` = 1.4, `y` = 0, `operacao` = "divisao"

````python
operacoes = {
     (111, 4.7, "soma"),
    (11, 44, "subtracao"),
    (71, 4, "multiplicacao"),
    (1.4, 0, "divisao")
}

for x, y, operacao in operacoes:
    if operacao == "soma":
        resultado = x + y
    elif operacao == "subtracao":
        resultado = x - y
    elif operacao == "multiplicacao":
        resultado = x * y
    elif operacao == "divisao":
        if y == 0:
            resultado = "erro"
        else:
            resultado = x / y

    print(f"{x} {operacao} {y} = {resultado}")
````

### **Checagem**:

````python
assert resultado_1 == 115.7
assert resultado_2 == -33.0
assert resultado_3 == 284.0
assert resultado_4 == "erro"
print("Exercício correto!")
````

## **Exercício 8: Pattern Matching**

Refaça o exercíco com match e compare a legibilidade de ambas soluções.

````python
# Entre com seu código aqui
operacoes = {
     (111, 4.7, "soma"),
    (11, 44, "subtracao"),
    (71, 4, "multiplicacao"),
    (1.4, 0, "divisao")
}

for x, y, operacao in operacoes:
  match operacao:
    case "soma":
      resultado = x + y
    case "subtracao":
      resultado = x - y
    case "multiplicacao":
      resultado = x * y
    case "divisao":
      resultado = 0 if y == 0 else x / y

  print(f"{x} {operacao} {y} = {resultado}")
````

### **Checagem**:

````python
assert resultado_1 == 115.7
assert resultado_2 == -33.0
assert resultado_3 == 284.0
assert resultado_4 == "erro"
print("Exercício correto!")
````

## **Exercício 9: prática de python - verificação da categoria do triângulo**

Verifique se os lados formam um triângulo:

* $𝑎 + 𝑏 > 𝑐$,
* $𝑎 + 𝑐 > 𝑏$,
* $𝑏 + 𝑐 > 𝑎$

Se não formar, a classificação será "nao triangulo"

Caso forme, ela poderá ser:

* `"equilatero"` → $a = b = c$,
* `"isosceles"` → $a = b$ ou $a = c$ ou $b = c$,
* `"escaleno"` → $a \neq b \neq c$

Classifique os seguintes triângulos:

Você deve implementar a lógica de um sistema simplificado de avaliação acadêmica, com base em critérios de desempenho e frequência.

Considere as seguintes variáveis de entrada:

* `prova`: nota da prova (valor entre 0 e 10)
* `trabalho`: nota do trabalho (valor entre 0 e 10)
* `frequencia`: percentual de presença (valor entre 0 e 100)
* `recuperacao`: nota da recuperação (valor entre 0 e 10), utilizada apenas se necessário

---

### **Etapa 1 – Validação dos dados**

Verifique se todos os valores estão dentro dos intervalos esperados. Caso **qualquer valor esteja fora do intervalo válido**, o resultado deve ser:

  ```python
  "invalido"
  ```

---

### **Etapa 2 – Cálculo da média inicial**

Se os dados forem válidos, calcule a média ponderada:

$\text{media} = 0.7 \cdot \text{prova} + 0.3 \cdot \text{trabalho}$

---

### **Etapa 3 – Verificação de frequência**

Se a frequência for menor que 75, o aluno está automaticamente reprovado, independentemente da média.

---

### **Etapa 4 – Classificação por desempenho**

Caso a frequência seja suficiente, classifique com base na média:

* `media ≥ 7` → `"aprovado"`
* `5 ≤ media < 7` → `"recuperacao"`
* `media < 5` → `"reprovado"`

---

### **Etapa 5 – Processo de recuperação**

Se o aluno estiver em recuperação:

1. Utilize a variável `recuperacao`
2. Calcule a nova média, como sendo a média entre a média anterior e a nota de recuperação.
3. Classifique o resultado final, sabendo que se a média final for maior que 6 o aluno será aprovado, caso contrário ele será reprovado.

---

Classifique uma turma onde o primeiro aluno tirou 8 na prova, 7 no trabalho e teve 80% de frequência, o segundo aluno tirou 9 tanto na prova, quanto no trabalho e teve 65% de frequência, um terceiro aluno tirou 5 tanto na prova, quanto no trabalho, com 90% de frequência, tirando 7 na recuperação e um quarto aluno tirou 5 na prova, 5 no trabalho, teve 95% de frequência e tirou 4.5 na recuperação. Um quinto aluno tirou 3 na prova, 4 no trabalho e teve 90% de frequência.

````python
# Entre com seu código aqui
alunos = [
    {"prova": 8, "trabalho": 7, "frequencia": 80, "recuperacao": None},
    {"prova": 9, "trabalho": 9, "frequencia": 65, "recuperacao": None},
    {"prova": 5, "trabalho": 5, "frequencia": 90, "recuperacao": 7},
    {"prova": 5, "trabalho": 5, "frequencia": 95, "recuperacao": 4.5},
    {"prova": 3, "trabalho": 4, "frequencia": 90, "recuperacao": None}
]

for i, aluno in enumerate(alunos):
  prova = aluno["prova"]
  trabalho = aluno["trabalho"]
  frequencia = aluno["frequencia"]
  recuperacao = aluno["recuperacao"]

  # validacao
  if not (0 <= prova <= 10 and 0 <= trabalho <= 10 and 0 <= frequencia <= 100 and
            (recuperacao is None or 0 <= recuperacao <= 10)):
        resultado = "invalido"

  else:
      # Média inicial
      media = 0.7 * prova + 0.3 * trabalho

      # Frequência
      if frequencia < 75:
          resultado = "reprovado"

      else:

          # Classificação

          if media >= 7:
              resultado = "aprovado"
          elif media >= 5:

              # Recuperação

              if recuperacao is not None:
                  media_final = (media + recuperacao) / 2
                  resultado = "aprovado" if media_final > 6 else "reprovado"
              else:
                  resultado = "recuperacao"
          else:
              resultado = "reprovado"

  print(f"Aluno {i+1}: {resultado}")
````

### **Checagem**:

````python
assert aluno_1 == "aprovado"
assert aluno_2 == "reprovado"
assert aluno_3 == "aprovado"
assert aluno_4 == "reprovado"
assert aluno_5 == "reprovado"

print("Exercício correto!")
````
