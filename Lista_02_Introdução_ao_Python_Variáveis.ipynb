## **Exercício 3: conversão de `float` para `int`**

Seja:

* `f1 = 3.1`
* `f2 = 3.33333`
* `f3 = 3.5`
* `f4 = 3.66666`
* `f5 = 3.99999`

Converta-as para inteiro e armazene os resultados nas variáveis `i1`, `i2`, `i3`,`i4` e `i5`

Avalie o resultado da conversão utilizando as seguintes funções:

* `int()`
* `round()`
* `round(XXX, 2)` (* onde XXX é a variável sendo arredondada)
*  `math.floor()` (* requer `import math`)
*  `math.ceil()` (* requer `import math`)

Observe os valores finais de cada variável e seus tipos utilizando a função `type()`.

````python
# Entre com seu código aqui
import math

f1 = 3.1
f2 = 3.33333
f3 = 3.5
f4 = 3.66666
f5 = 3.99999

i1 = (int(f1))
i2 = (int(f2))
i3 = (int(f3))
i4 = (int(f4))
i5 = (int(f5))

print("int():", i1, i2, i3, i4, type(i1))
print("round():", round(f1), round(f2), round(f3), round(f4), round(f5), type(round(f1)))
print("round(xxx, 2):", round(f1, 2), round(f2, 2), round(f3, 2), round(f4, 2), round(f5, 2), type(round(f1, 2)))
print("math.floor():", math.floor(f1), math.floor(f2), math.floor(f3), math.floor(f4), math.floor(f5), type(math.floor(f1)))
print("math.ceil():", math.ceil(f1), math.ceil(f2), math.ceil(f3), math.ceil(f4), math.ceil(f5), type(math.ceil(f1)))

````

### **Checagem**:

````python
assert isinstance(i1, int)
assert isinstance(i2, int)
assert isinstance(i3, int)
assert isinstance(i4, int)
assert isinstance(i5, int)
print("Exercício correto!")
````

## **Exercício 8: prática de python - cálculo de potência de um motor**

Considere que um sistema de monitoramento de um motor elétrico recebe alguns dados em formato de texto provenientes sensores:

* `corrente = "12.5"`
* `tensao = "220"`
* `rpm = "1450"`

Esses valores representam:

* A corrente elétrica em ampères
* A tensão elétrica em volts
* A velocidade do motor em rpm

Tarefas:

1. Converta as variáveis corrente e tensao para valores numéricos adequados;
2. Calcule a potência elétrica aproximada do motor usando a fórmula:

 <div align="center"> P = V x I </div>

onde:

* P é a potência em watts
* V é a tensão
* I é a corrente

3. Converta rpm para número inteiro e calcule a nova rotação após um aumento de 300 rpm.
4. Armazene o valor final da rotação na variável: `nova_rpm`

5. Crie uma string chamada `relatorio` com o seguinte formato:

```
Motor em operação
Rotação atual: XXXX rpm
Nova rotação: YYYY rpm
Potência elétrica estimada: ZZZZ W
```
onde:

* XXXX corresponde ao valor original de rotação
* YYYY corresponde à nova rotação
* ZZZZ corresponde à potência calculada

6. Imprima o valor da variável `relatorio`.


````python
# Entre com seu código aqui
corrente = "12.5"
tensao  = "220"
rpm = "1450"

correntec = (float(corrente))
tensaoc = (int(tensao))
rpmc = (int(rpm))

pontecia = correntec * tensaoc
nova_rpm = rpmc + 300

relatorio = f"Motor em operação\nRotação atual: {rpmc} rpm\nNova rotação: {nova_rpm} rpm\nPotência elétrica estimada: {int(pontecia)} W"
````

### **Checagem**:

````python
assert relatorio == "Motor em operação\nRotação atual: 1450 rpm\nNova rotação: 1750 rpm\nPotência elétrica estimada: 2750 W"
print("Exercício correto!")
````

## **Exercício 9: prática de pytnon - estimativa de velocidade de uma esteira transportadora**

Um sistema de automação mede o deslocamento de uma caixa em uma esteira transportadora. Os dados recebidos pelo controlador chegam em formato de texto:

* `distancia = "2.75"`
* `tempo = "4.2"`

onde:

* `distancia` representa o deslocamento em metros
* `tempo` representa o tempo em segundos

O sistema também recebe o número de caixas transportadas por minuto, também em texto:

* `caixas_por_minuto = "18.6"`

Tarefas

1. Converta as variáveis `distancia` e `tempo` para valores numéricos;
2. Calcule a velocidade média da esteira usando a fórmula: v = t/d;
onde d é a distância (m) e t é o tempo (s). Armazene o resultado na variável `velocidade`

3. Arredonde o valor da velocidade para duas casas decimais e armazene em `velocidade_arredondada`

4. Converta caixas_por_minuto para valor numérico. Em seguida arredonde esse valor para o inteiro mais próximo. Armazene em `caixas_arredondadas`

5. Calcule o número estimado de caixas transportadas em 10 minutos. Armazene o resultado em: `total_caixas`

6. Calcule uma estimativa inteira do total de caixas, removendo a parte decimal sem arredondar (truncamento). Armazene em: `total_caixas_truncado`

7. Crie uma string chamada `relatorio` com o seguinte formato:

```
Relatório da esteira

Distância percorrida: X m
Tempo medido: Y s
Velocidade média: Z m/s

Caixas por minuto (sensor): A
Caixas por minuto (ajustado): B

Caixas estimadas em 10 minutos: C
Caixas estimadas (inteiro): D
```


````python
# Entre com seu código aqui
distancia = "2.75"
tempo = "4.2"
caixas_por_minuto = "18.6"

distancia = float(distancia)
tempo = float(tempo)
caixas_por_minuto = float(caixas_por_minuto)

velocidade = tempo / distancia
velocidade_arredondada = round(velocidade, 2)
caixas_arredondadas = round(caixas_por_minuto)

total_caixas = caixas_por_minuto * 10
total_caixas_truncado = int(total_caixas)

relatorio = f"Relatório da esteira\n\nDistância percorrida: {distancia} m\nTempo medido: {tempo} s\nVelocidade média: {velocidade_arredondada} m/s\n\nCaixas por minuto (sensor): {caixas_por_minuto}\nCaixas por minuto (ajustado): {caixas_arredondadas}\n\nCaixas estimadas em 10 minutos: {total_caixas}\nCaixas estimadas (inteiro): {total_caixas_truncado}"
````

### **Checagem**:

````python
import math

# verificar se variáveis existem
variaveis_necessarias = [
    "distancia",
    "tempo",
    "velocidade",
    "velocidade_arredondada",
    "caixas_por_minuto",
    "caixas_arredondadas",
    "total_caixas",
    "total_caixas_truncado",
    "relatorio"
]

for v in variaveis_necessarias:
    assert v in globals(), f"A variável '{v}' não foi criada."

# valores esperados
distancia_esperada = 2.75
tempo_esperado = 4.2

# CORRIGIDO: Velocidade calculada como tempo / distancia, conforme descrição do exercício
velocidade_esperada = tempo_esperado / distancia_esperada # Correção: alterado 'tempo_esperada' para 'tempo_esperado'
velocidade_arredondada_esperada = round(velocidade_esperada, 2)

caixas_por_minuto_esperado = 18.6
caixas_arredondadas_esperado = round(caixas_por_minuto_esperado)

# total_caixas_esperado deve usar o valor float original de caixas_por_minuto
total_caixas_esperado = caixas_por_minuto_esperado * 10
total_caixas_truncado_esperado = math.trunc(total_caixas_esperado)

# CORRIGIDO: relatorio_esperado com a nova velocidade e total_caixas
relatorio_esperado = (
    "Relatório da esteira\n\n"
    "Distância percorrida: 2.75 m\n"
    "Tempo medido: 4.2 s\n"
    "Velocidade média: 1.53 m/s\n\n" # Atualizado de 0.65 para 1.53
    "Caixas por minuto (sensor): 18.6\n"
    "Caixas por minuto (ajustado): 19\n\n"
    "Caixas estimadas em 10 minutos: 186.0\n"
    "Caixas estimadas (inteiro): 186"
)

# checagem de valores
assert distancia == distancia_esperada
assert tempo == tempo_esperado
assert abs(velocidade - velocidade_esperada) < 1e-9
assert velocidade_arredondada == velocidade_arredondada_esperada
assert caixas_por_minuto == caixas_por_minuto_esperado
assert caixas_arredondadas == caixas_arredondadas_esperado
assert total_caixas == total_caixas_esperado
assert total_caixas_truncado == total_caixas_truncado_esperado
assert relatorio == relatorio_esperado

# checagem de tipos
assert isinstance(distancia, float)
assert isinstance(tempo, float)
assert isinstance(velocidade, float)
assert isinstance(velocidade_arredondada, float)
assert isinstance(caixas_por_minuto, float)
assert isinstance(caixas_arredondadas, int)
assert isinstance(total_caixas, float)
assert isinstance(total_caixas_truncado, int)
assert isinstance(relatorio, str)

print("Exercício correto!")
````

## **Exercício 10 : prática de python - monitoramento de enchimento de um tanque**

Um sistema de automação industrial monitora o enchimento de um tanque
de armazenamento de líquido.
Os sensores enviam os dados ao controlador em formato de texto:

* `volume = "3.8"`
* `tempo = "2.5"`

onde:
* `volume` representa o volume transferido para o tanque em
litros (L)
* `tempo` representa o tempo da medição em segundos (s)

O sistema também recebe a vazão da bomba em litros por minuto,
também em formato de texto:
* `vazao_bomba = "22.7"`


### Tarefas:

1. Converta as variáveis `volume` e `tempo` para valores numéricos.
2. Calcule a vazão média observada usando a fórmula Q = V / t
onde: Q é a vazão (L/s), V é o volume transferido (L) e t é o tempo (s). Armazene o resultado na variável: `vazao_media`

3. Arredonde o valor da vazão média para duas casas decimais e armazene
em: `vazao_arredondada`

4. Converta `vazao_bomba` para valor numérico. Em seguida, arredonde esse valor para o inteiro mais próximo e armazene em:  `vazao_ajustada`

5. Calcule o volume estimado transferido em 10 minutos, considerando a
vazão da bomba. Armazene o resultado em: `volume_total`

6. Calcule uma estimativa inteira do volume transferido, removendo a
parte decimal sem arredondar (truncamento). Armazene em: `volume_truncado`

7. Crie uma string chamada `relatorio` com o seguinte formato:

```
    Relatório do tanque

    Volume medido: X L
    Tempo de medição: Y s
    Vazão média observada: Z L/s

    Vazão da bomba (sensor): A L/min
    Vazão da bomba (ajustada): B L/min

    Volume estimado em 10 minutos: C L
    Volume estimado (inteiro): D L
```

````python
# Entre com seu código aqui
volume = "3.8"
tempo = "2.5"
vazao_bomba = "22.7"

volume = float(volume)
tempo = float(tempo)
vazao_bomba = float(vazao_bomba)

vazao_media = volume / tempo
vazao_arredondada = round(vazao_media, 2)
vazao_ajustada = round(vazao_bomba)

volume_total = vazao_bomba * 10
volume_truncado = int(volume_total)

relatorio = (
    "Relatório do tanque\n\n"
    f"Volume medido: {volume} L\n"
    f"Tempo de medição: {tempo} s\n"
    f"Vazão média observada: {vazao_arredondada} L/s\n\n"
    f"Vazão da bomba (sensor): {vazao_bomba} L/min\n"
    f"Vazão da bomba (ajustada): {vazao_ajustada} L/min\n\n"
    f"Volume estimado em 10 minutos: {volume_total} L\n"
    f"Volume estimado (inteiro): {volume_truncado} L"
)
````

### **Checagem**:

````python
import math

# Verificar se todas as variáveis necessárias foram criadas
variaveis_necessarias = [
    "volume",
    "tempo",
    "vazao_media",
    "vazao_arredondada",
    "vazao_bomba",
    "vazao_ajustada",
    "volume_total",
    "volume_truncado",
    "relatorio"
]

for v in variaveis_necessarias:
    assert v in globals(), f"A variável '{v}' não foi criada."

# Valores esperados
volume_esperado = 3.8
tempo_esperado = 2.5
vazao_media_esperada = volume_esperado / tempo_esperado
vazao_arredondada_esperada = round(vazao_media_esperada, 2)

vazao_bomba_esperada = 22.7
vazao_ajustada_esperada = round(vazao_bomba_esperada)

volume_total_esperado = vazao_bomba_esperada * 10
volume_truncado_esperado = math.trunc(volume_total_esperado)

relatorio_esperado = (
    "Relatório do tanque\n\n"
    "Volume medido: 3.8 L\n"
    "Tempo de medição: 2.5 s\n"
    "Vazão média observada: 1.52 L/s\n\n"
    "Vazão da bomba (sensor): 22.7 L/min\n"
    "Vazão da bomba (ajustada): 23 L/min\n\n"
    "Volume estimado em 10 minutos: 227.0 L\n"
    "Volume estimado (inteiro): 227 L"
)

# Checagem de valores
assert volume == volume_esperado
assert tempo == tempo_esperado
assert abs(vazao_media - vazao_media_esperada) < 1e-9
assert vazao_arredondada == vazao_arredondada_esperada
assert vazao_bomba == vazao_bomba_esperada
assert vazao_ajustada == vazao_ajustada_esperada
assert volume_total == volume_total_esperado
assert volume_truncado == volume_truncado_esperado
assert relatorio == relatorio_esperado

# Checagem de tipos
assert isinstance(volume, float)
assert isinstance(tempo, float)
assert isinstance(vazao_media, float)
assert isinstance(vazao_arredondada, float)
assert isinstance(vazao_bomba, float)
assert isinstance(vazao_ajustada, int)
assert isinstance(volume_total, float)
assert isinstance(volume_truncado, int)
assert isinstance(relatorio, str)

print("Exercício correto!")
````
